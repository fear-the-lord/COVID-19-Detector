```python
#Import necessary packages 

import pandas as pd
import os 
import shutil
import random
```

```python
#Create the covid positive dataset 

FILE_PATH = "covid-chestxray-dataset-master/metadata.csv"
IMAGE_PATH = "covid-chestxray-dataset-master/images"
```

```python
#Create a data frame for the .csv file 

frame = pd.read_csv(FILE_PATH)
print(frame.shape)
```

```python
#Create a new directory that will store the covid images 

TARGET_DIR = "Dataset/Covid"

if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Covid Folder Created")
```

```python
# Now pick images from the dataset and place it in the covid folder
count = 0 # counts the no:of covid-19 images 

# Some x-rays have a front view, some x-rays have top or side view. We only need the x-rays with top view 

for (i, row) in frame.iterrows():
    if row["finding"] == "COVID-19" and row["view"] == "PA":
        filename = row["filename"]
        image_path = os.path.join(IMAGE_PATH , filename)
        image_copy_path = os.path.join(TARGET_DIR, filename)
        shutil.copy2(image_path, image_copy_path)
        print("Moving Image ", count)
        count += 1
```

```python
# Moving exact no:of images from the kaggle dataset to our non-covid folder 

KAGGLE_PATH = "chest_xray/train/NORMAL"
TARGET_PATH = "Dataset/Non-Covid"

image_names = os.listdir(KAGGLE_PATH)
random.shuffle(image_names)

for i in range(count):
    image_name = image_names[i]
    image_path = os.path.join(KAGGLE_PATH, image_name)
    image_copy_path = os.path.join(TARGET_PATH, image_name)
    shutil.copy2(image_path, image_copy_path)
    print("Moving image ", i)
```
